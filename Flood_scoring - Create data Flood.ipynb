```python
data_drias = pd.read_csv(r"data\Drias_data\RCP_4.5.csv", sep=";", header=31)
```

```python
# Load flood data
Flood_file = {
    "high": ["n_iso_ht_01_01for_s.shp", "n_iso_ht_02_01for_s.shp", "n_iso_ht_03_01for_s.shp"],
    "mid": ["n_iso_ht_01_02moy_s.shp", "n_iso_ht_02_02moy_s.shp", "n_iso_ht_03_02moy_s.shp"],
    "low": ["n_iso_ht_01_04fai_s.shp", "n_iso_ht_02_04fai_s.shp", "n_iso_ht_03_04fai_s.shp"]
}
```

```python
print("Loading flood data...")
all_flood_data = []

for scenario, files in Flood_file.items():
    scenario_data = []
    for file in files:
        full_link = r'data\map\tri_2020_sig_di\\' + file
        gdf = gpd.read_file(full_link)
        gdf = gdf[['geometry', 'ht_min', 'ht_max']]
        scenario_data.append(gdf)
    
    scenario_gdf = pd.concat(scenario_data, ignore_index=True)
    scenario_gdf['scenario'] = scenario
    all_flood_data.append(scenario_gdf)

flood_gdf = pd.concat(all_flood_data, ignore_index=True)

print(f"Flood data loaded: {len(flood_gdf)} features worldwide")
print(f"CRS: {flood_gdf.crs}")
```

# Create points

```python
# Create points FIRST
print("\nCreating point geometries...")
points_gdf = gpd.GeoDataFrame(
    data_drias,
    geometry=[Point(lon, lat) for lon, lat in zip(data_drias['Longitude'], data_drias['Latitude'])],
    crs=flood_gdf.crs
)

print(f"Points created: {len(points_gdf)}")
```

# Filter points that doesn't belong to metropolitan France

```python
# === CRITICAL FIX 1: Filter flood data to only your region ===
print("\nFiltering flood data to only France region...")

# Get bounds of your points with a buffer
point_bounds = points_gdf.total_bounds
buffer_margin = 0.5  # degrees (about 50km)

min_x = point_bounds[0] - buffer_margin
min_y = point_bounds[1] - buffer_margin
max_x = point_bounds[2] + buffer_margin
max_y = point_bounds[3] + buffer_margin

print(f"Point extent: X=[{point_bounds[0]:.2f}, {point_bounds[2]:.2f}], Y=[{point_bounds[1]:.2f}, {point_bounds[3]:.2f}]")
print(f"Filtering to: X=[{min_x:.2f}, {max_x:.2f}], Y=[{min_y:.2f}, {max_y:.2f}]")

# Filter flood data to this region
flood_gdf = flood_gdf.cx[min_x:max_x, min_y:max_y]

print(f"Filtered flood data: {len(flood_gdf)} features (reduced by {(1 - len(flood_gdf)/4154115)*100:.1f}%)")
```

# Create spatial index and create radius for buffer

```python
# === CRITICAL FIX 2: Convert buffer_distance to degrees ===
# 10 meters in degrees at France latitude (~45°)
# At 45° latitude: 1 degree longitude ≈ 78 km, 1 degree latitude ≈ 111 km
# So 10 meters ≈ 0.0001 degrees

buffer_distance_meters = 2500
buffer_distance_degrees = buffer_distance_meters / 111000  # Rough conversion

print(f"\nBuffer distance: {buffer_distance_meters}m = {buffer_distance_degrees:.6f} degrees")

# Create spatial index on filtered data
print("Creating spatial index on filtered data...")
sindex = flood_gdf.sindex
print("Spatial index created!")
```

# Join drias and flood dataframe spatially

```python
# === Process points ===
buffer_distance = buffer_distance_degrees
results = []

start_time = time.time()
print(f"\nProcessing {len(points_gdf)} points...")

for idx_num, (idx, point_row) in enumerate(points_gdf.iterrows()):
    
    # Progress update every 100 points
    if idx_num % 100 == 0:
        elapsed = time.time() - start_time
        if idx_num > 0:
            rate = idx_num / elapsed
            remaining = len(points_gdf) - idx_num
            est_remaining = remaining / rate / 60
            
            print(f"Point {idx_num}/{len(points_gdf)} ({idx_num/len(points_gdf)*100:.1f}%) | "
                  f"Rate: {rate:.2f} pts/sec | Elapsed: {elapsed:.1f}s | "
                  f"Est. remaining: {est_remaining:.1f}min")
    
    point = point_row.geometry
    lon = point_row['Longitude']
    lat = point_row['Latitude']
    
    # Create bounding box
    x, y = point.x, point.y
    bbox = box(x - buffer_distance, y - buffer_distance, 
               x + buffer_distance, y + buffer_distance)
    
    # Query spatial index
    possible_matches_idx = list(sindex.query(bbox))
    
    # Debug first point
    if idx_num == 0:
        print(f"\nFirst point found {len(possible_matches_idx)} candidates (should be < 100)")
    
    if not possible_matches_idx:
        results.append({
            'longitude': lon,
            'latitude': lat,
            'scenario': [],
            'ht': {'high': [], 'mid': [], 'low': []}
        })
        continue
    
    # Get candidates
    possible_matches = flood_gdf.iloc[possible_matches_idx]
    
    # Check distances
    point_data = {'scenarios': set(), 'ht_data': defaultdict(list)}
    
    for _, flood_row in possible_matches.iterrows():
        if point.distance(flood_row.geometry) <= buffer_distance:
            scenario = flood_row['scenario']
            point_data['scenarios'].add(scenario)
            point_data['ht_data'][scenario].append([flood_row['ht_min'], flood_row['ht_max']])
    
    # Build result
    if point_data['scenarios']:
        scenarios = list(point_data['scenarios'])
        ht_dict = {
            'high': point_data['ht_data'].get('high', []),
            'mid': point_data['ht_data'].get('mid', []),
            'low': point_data['ht_data'].get('low', [])
        }
    else:
        scenarios = []
        ht_dict = {'high': [], 'mid': [], 'low': []}
    
    results.append({
        'longitude': lon,
        'latitude': lat,
        'scenario': scenarios,
        'ht': ht_dict
    })

result_df = pd.DataFrame(results)

# === Summary ===
total_time = (time.time() - start_time) / 60

print("\n" + "="*60)
print("COMPLETE!")
print("="*60)
print(f"Total processing time: {total_time:.1f} minutes")
print(f"Total points processed: {len(result_df)}")
print(f"Points with flood risk: {len(result_df[result_df['scenario'].apply(len) > 0])}")
print(f"Points with no flood risk: {len(result_df[result_df['scenario'].apply(len) == 0])}")

print("\nFirst few rows:")
print(result_df.head())

print("\nExample of point WITH flood risk:")
with_risk = result_df[result_df['scenario'].apply(len) > 0]
if len(with_risk) > 0:
    print(with_risk.head(1))
else:
    print("No points found with flood risk")

# Save results
#print("\nSaving results...")
#result_df.to_pickle('data\Flood\flood_risk_results.pkl')
#print("Results saved to 'data\Flood\flood_risk_results.pkl'")
```
